```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import keras
import keras.backend as K
from keras.applications.inception_v3 import inception_v3
from keras.applications.vgg16 import VGG16

from skimage.io import imread, imshow
from skimage.transform import resize
```

```python
#prepare the model (with default imagenet weights)
K.clear_session()
#model = inception_v3.InceptionV3(input_shape=(299,299,3),weights="imagenet")
print("load model")
model = VGG16(input_shape=(224,224,3),weights="imagenet")
```

```python
from kwb.basic_fn import show_inference, first_conv_filters, intermediate_output
#keras basic tools
```

```python
#image obtained from wikipedia
image_path = "resources/images/-_panoramio_-_Basa_Roland-2.jpg"

#read and process image
_x = imread(image_path)
x = resize(_x,(224,224))*255.
x  =x.astype(np.uint8)
```

```python
#show image
plt.imshow(x)
plt.show()
#show inference result
show_inference(model,x)
```

# occulusion

occlusion analysis is finding the important pixel within the image by creating numbers of images with each pixel nullyfied and check how the output changed. if 224 x 224 images, 224 x 224 images have to be prepared and analyzed so it is very costly and not preferred. so sacrificing the final output resolution, we will occulude every XX pixel by the argv skip_pixel and forcely put back to the original resolution by resizing. another way of occulusion is enlarging the occuluding area (ex. 4 x 4) but makes the output shape complicated not writing the codes for now.

```python
from kwb.attention_map import pixel_wise_occlusion
```

```python
vals = pixel_wise_occlusion(model,x,skip_pixel=32)
plt.imshow(x)
plt.imshow(vals,alpha=0.5,cmap="jet")
plt.show()

vals = pixel_wise_occlusion(model,x,skip_pixel=16)
plt.imshow(x)
plt.imshow(vals,alpha=0.5,cmap="jet")
plt.show()

vals = pixel_wise_occlusion(model,x,skip_pixel=8)
plt.imshow(x)
plt.imshow(vals,alpha=0.5,cmap="jet")
plt.show()
```

# vanilla backpropagation

```python
from kwb.attention_map import saliency_map

vb = saliency_map(model,x,"predictions",normalize="img_standard")
plt.imshow(vb)
plt.title("with standard deprocessed gradients")
plt.show()
```

lets try different methods to visualize important pixels
# guided back propagation

```python
from kwb.model_modification import linearize_model,build_guided_model
lmodel = linearize_model(model)
gmodel = build_guided_model(model)
```

```python
#close to raw output but normalized for image visualization
gb = saliency_map(gmodel,x,"predictions",normalize="img_standard")
plt.imshow(gb)
plt.title("with standard deprocessed gradients")
plt.show()

#absolute gradients
gb = saliency_map(gmodel,x,"predictions",normalize="abs")
plt.imshow(gb,cmap="jet")
plt.title("with absolute gradients")
plt.show()

#positive gradients. meaning that nullyfying the highlighted color pixels reduce the classification accuracy
gb = saliency_map(gmodel,x,"predictions",normalize="pos")
plt.title("positive gradients")
plt.imshow(gb)
plt.show()

#positive gradients. meaning that coloring the respective pixel with the indicated color reduce the classfication accuracy
gb = saliency_map(gmodel,x,"predictions",normalize="neg")
plt.title("negative gradients")
plt.imshow(gb)
plt.show()
```

# batch analysis
Most of the functions above are compatible with batch inputs.
incase of feeding batch instead of single input, the returning values will be added a batch demention to the first element.


```python
image_path = ["resources/images/-_panoramio_-_Basa_Roland-2.jpg","resources/images/Kittyply_edit1.jpg"]

images = []
for p in image_path:
    _x = imread(p)
    _x = resize(_x,(224,224))*255.
    images.append(_x.astype(np.uint8))
images = np.array(images)
print(images.shape)
```

```python
for image in images:
    plt.imshow(image)
    plt.show()
```

```python
intout = intermediate_output(model,images,"block1_conv1",visualize=True,
                             row=6,col=6,cmap="jet")
print(intout.shape) 
#note that feeding single image (224,224,3) returns (224,224,64), but feeding batch (2,224,224,3) returns (2,224,224,64)
```
