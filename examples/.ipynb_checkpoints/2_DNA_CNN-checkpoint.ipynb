```python
import kwb.utils
import kwb.dna_utils
import kwb.models

import keras.backend as K
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
K.clear_session()
# sample dna network that discriminates a 2000bp sequence has a specific feature (1) or not (0). 
model = kwb.models.build_dna_model()
#linearize last activation layer and modify relu to guided relu.
gmodel = kwb.utils.build_guided_model(model)
model.summary()
```

##### input information 

the input shape follows tensorflow order

(1,1,2000,4) 
(batch_size, H, W, channel )

    H: always 1. did not delete this dimention so that all the utils for images can be shared with dna models.

    W: bp. 2000.

    channel: ACGT 4

    values. after one-hot encoding, 0.25 is subtracted so that the range will be between -0.25 to 0.75

```python
# sample sequence.
x = np.array("GGCGCCCTTCGACCAAGGGACCTAAGTAATGGAAGGGGAAAATCCAAATGTGAAAATTATCCCGGATCACATAGCTAGTAAAGTAGTAATATGTCTTATAAACCAACAACTCATATTGTTTTTATGGTTGTACAATAGAGAACACGGTAAAAGCATCGGCACATTTTACAAAGGCTTCCGCATGTGCTCCTCGGTGATAATAGCATAACACATCTGAAGCAGGGGTAGAACCGTGGCTTTAGCAGCTTGGGCCCGAGCCCTACTCGTGGAAGAAATTATCATATTTTCAGACTTTCTCAACATCGCTTGGCCCAACCCTACTCCTCTAGCCCAGCCCCACGCACATCCTTTCCCAACCTCAAGTGGCCAGGTTCCTAGCCTAATAGACGGATCACAGCCCAATAGACAACCCAAGTGACCAACCCTAATTCAGTGTCCTCACTCCTCACTTCCCAATAGACAACCCTAAACTATAATCTCCACCAGTCTACCGTCCTCACTCCTCACGCTGCACTCATGACTCCAACGTTGCATCATGCATCGGTGCATGTTCACTGCTCCTCTCATCTCGCAACATCGGCTGGGGCGCGCATCGATAGCGGCAACAACTTGGATCTGCATTCTCGTTTTGATCCCCCAAATCCAGGTTAGTCTAGTCTCTAATCTTTGATGTACATTCTCATTTTGTATTTGGTTGTAGACAAGCAGTTAGTGAGAAATTTGGTTTGATTTCCTAGATAAAGAGGAGCAGAACAAATATAGGGGTGAATGGGAGCTTTCGGAAGTTATGTTTTAAAAAGCACAATTGGTTGCTAAGGATGCAATTTTCTGTTTTAATTGCTTCCTTTCTAAGCAAAACATGAGATGTTTTTGTTGTTTGAAGAGCGAGATAATTTTGCTAGTGTCGAAGATATTACAGAACATTATCATAATCGTGCTTTGAGAAATGAAGATCATCGAGGTATGTCATATATTCCTTGTAAATAGTTAAGTTCTCATAATTTGCAGTAATATCATTTTTCAGATTCAAGTTCAGCCTCTCTCCATACTGTAGTGCTGTTGATGGAGAGTGAAGATCATGGGGGTATTCCAATGAAATGTTAATTCAGTTCTTTTTAATTTTTACTGTACTTCCGTTCCGTGTATGAATTGTCATGGTAATTTTAGCCCTCATTTTTGCTTTTTTTTTGACATTTTTACTGTACTTGGTCTATGTGTGGTCATGGGCAATTTTAATCTTCAATTTGAGCTCTAGGCTAGAAGAAATCCTAACGCCACCACTAATCTGAAGCACGAAACAAATCAATTAAAGAAGCCATAAATGATAAATGAATAATCCTTCATGCATAAAAATATTTAGTTATCCATAAAGAGACGACAGACGTATGCAACGTTCATATTAATTGTCACTCTACAGTGTCACCGAGTTGTCGGTCCAAGCAATTTTACCCATACATGAAAACCATCTGAGAAAGGGCATGAGAGTAGTACTGCGTCAATACACATGCAGTTGCACACAGTAGCTGAGAGAGCGAGCCGGAAACATCTGGTCGCCGGTCCACGATCGGTCTACGCATGGGCTTCCCCCCGACCAGGTGGTGCGTGCGAGCCTACCACATGTCCACATGCATGCACGGTGATCGACATCCCGATCGACGTGTCGACAGTGCAGCCACCCCGCCGCTCGCTCGCTCGCGCTCGCGCAGTCGCGCTAGCACGGACATACGGAGTAGTATATGCGTATGCATAGGCACGCGCGCGCACGCGCATGCAGCGAACCAGAGCGCGCCGCGCACGGGCGGAGGCATAGCTTGCGCACGTACGTGTCGCGCTCCGGCGGCCTCGCCACGACGCGTGTCGCGGAGGCGCGCGCGCGCACCGAGCCGTCCGCGTCGCCGCGCGACACGTACGTTGCCGGCGCCCTGATCCCGCCGCGGAGCACCGCCCCGTCCGTCGCCGCCTATAAGAAGGACAGCGCGGCCGCGTCGTCCATCCCCATC")
x = kwb.dna_utils.one_hot(np.expand_dims(x,axis=0))-0.25

print(x.shape)
print(model.predict(x)) #0.964 meaning predicted as a promoter
print(gmodel.predict(x)) #3.31 linearized the last layer, so outputing over 1.
```

##### visualize the importance with guided gradcam

gradcam uses the layer penultimate to the flatten or global average pooling layer.

```python
gc = kwb.utils.grad_cam(gmodel,x,"max_pooling2d_6",method="naive")
gcpp = kwb.utils.grad_cam(gmodel,x,"max_pooling2d_6",method="gradcampp")

print(gc.shape,gcpp.shape) #(1,1,2000) (batch,H,W)

plt.plot(np.arange(2000),gc[0,:],label="naive")
plt.plot(np.arange(2000),gcpp[0,:],label="gradcampp")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

##### visualize the bp wise importance with guided backpropagation. uses the last layer modified to linear.

can use the preprediction layer as seen in the summary if the activation layer is isolated from the dense layer.

raw values

```python
gb = kwb.utils.guided_backprop(gmodel,x,"predictions",normalize="raw")
print(gb.shape) #(batch,H,W,channel)
plt.title("raw")
for i,n in zip(range(x.shape[3]),["A","C","G","T"]): #repeat per channel
    plt.plot(np.arange(2000),gb[0,:,i],label=n)
```

absolute gradients 

```python
gb = kwb.utils.guided_backprop(gmodel,x,"predictions",normalize="abs")
print(gb.shape) #(batch,H,W)
plt.title("absolute gradients")
plt.plot(np.arange(2000),gb[0,:])
```

focusing on only positive gradients

meaning that nullifying the values lowers the prediction value.

```python
gb = kwb.utils.guided_backprop(gmodel,x,"predictions",normalize="pos")
print(gb.shape)
plt.title("positive gradients")
for i,n in zip(range(x.shape[3]),["A","C","G","T"]): #repeat per channel
    plt.plot(np.arange(2000),gb[0,:,i],label=n)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

negative gradient means that substituting it to the indicating values lowers the prediction value.

note that the result is nearly complementary to the pos result.

if substituting A or T or C in certain position lowers prediction, means same to that nullyfing G lowers prediction.

```python
gb = kwb.utils.guided_backprop(gmodel,x,"predictions",normalize="neg")
print(gb.shape)
plt.title("negative gradients")
for i,n in zip(range(x.shape[3]),["A","C","G","T"]): #repeat per channel
    plt.plot(np.arange(2000),gb[0,:,i],label=n)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

##### pixel wise occulusion. replace 1bp (1pixel) with a fixed value and see the effect.

in default, will replace to [-0.25,-0.25,-0.25,-0.25] which is a "zero hot" which does not exist.

batch compatible as in gradcam and backpropagation but watch out for memory usage

```python
#different compared to images, even single nucleotide occulusion generates only 2000 sequence, so skipping nothing. skip_pixel = 1
oc = kwb.utils.pixel_wise_occlusion(model,x,skip_pixel=1)
print(oc.shape) #(batch,H,W)
plt.plot(oc[0,:])
```

```python
#recalculate gc and gb and compare together with occlusion
gc = kwb.utils.grad_cam(gmodel,x,"max_pooling2d_6",method="naive")
gb = kwb.utils.guided_backprop(gmodel,x,"predictions",normalize="abs")

plt.plot(np.arange(2000),oc[0,:],label="pixel-wise occulusion",alpha=0.3)
plt.plot(np.arange(2000),gc[0,:],label="guided grad-cam")
plt.plot(np.arange(2000),gb[0,:],label="guided backpropagation")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Nucleotides important")
plt.show()
```
