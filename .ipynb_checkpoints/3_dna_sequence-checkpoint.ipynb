focusing on only positive gradients

meaning that nullifying the values lowers the prediction value.

```python
gb = guided_backprop(gmodel,x,"predictions",normalize="pos")
print(gb.shape)
plt.title("positive gradients")
for i,n in zip(range(x.shape[3]),["A","C","G","T"]): #repeat per channel
    plt.plot(np.arange(2000),gb[0,:,i],label=n)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

negative gradient means that substituting it to the indicating values lowers the prediction value.

note that the result is nearly complementary to the pos result.

if substituting A or T or C in certain position lowers prediction, means same to that nullyfing G lowers prediction.

```python
gb = guided_backprop(gmodel,x,"predictions",normalize="neg")
print(gb.shape)
plt.title("negative gradients")
for i,n in zip(range(x.shape[3]),["A","C","G","T"]): #repeat per channel
    plt.plot(np.arange(2000),gb[0,0,:,i],label=n)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
def pixel_wise_occlusion(model, images,substitute_values=-0.25,method="normalize"):
    #substitute pixel to certain value and infer prediction.
    if len(images.shape) == 3:
        images = np.expand_dims(images,axis=0)
    B = images.shape[0]
    H = images.shape[1]
    W = images.shape[2]
    C = images.shape[3]
    vals = []
    #infer per image
    for batch in range(B): #iterate per image
        substituted = []
        base_val = model.predict(np.expand_dims(images[batch],axis=0))[0]
        for h in range(H):
            for w in range(W):
                tmp = images[batch].copy() #1,2000,4
                replace_val = np.array([substitute_values]*C)
                tmp[h,w] = replace_val
                substituted.append(tmp)     
        substituted = np.array(substituted)
        substituted = np.reshape(substituted,(H*W,H,W,C))
        
        val = model.predict(substituted,batch_size=32)
        if method=="normalize":
            #normalize by subtracting the default inference value
            val -= base_val
            #negative values means important so swaping pos neg
            val *= -1
            val = np.maximum(0,val)
            val /= val.max()
        vals.append(val)
    
    vals = np.array(vals)
    vals = np.reshape(vals,(B,H,W))
    return vals
```

##### pixel wise occulusion. replace 1bp (1pixel) with a fixed value and see the effect.

in default, will replace to [-0.25,-0.25,-0.25,-0.25] which is a "zero hot" which does not exist.

batch compatible as in gradcam and backpropagation but watch out for memory usage

```python
#will use linear model so that inference will not be cutoffed by 1.0
lmodel = linearize_activation(model)
```

```python
oc = pixel_wise_occlusion(lmodel,x)
print(oc.shape) #(batch,H,W)
plt.plot(oc[0,0,:])
```

```python
#lets compare with different methods
gc = grad_cam(gmodel,x,"max_pooling2d_6",method="naive")
gb = guided_backprop(gmodel,x,"predictions",normalize="abs")
oc = pixel_wise_occlusion(lmodel,x)
```

```python
plt.plot(np.arange(2000),oc[0,0,:],label="pixel-wise occulusion",alpha=0.3)
plt.plot(np.arange(2000),gc[0,0,:],label="guided grad-cam")
plt.plot(np.arange(2000),gb[0,0,:],label="guided backpropagation")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Nucleotides important")
plt.show()
```
