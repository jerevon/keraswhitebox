```python
from utils import *

import keras
from keras.applications.inception_v3 import inception_v3
from keras.applications.vgg16 import VGG16
from skimage.io import imread, imshow
from skimage.transform import resize
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#prepare the model (with default imagenet weights)
K.clear_session()
#model = inception_v3.InceptionV3(input_shape=(299,299,3),weights="imagenet")
print("load model")
model = VGG16(input_shape=(224,224,3),weights="imagenet")
```

```python
#image from wikipedia
image_path = "images/Golden_Retriever_Carlos_(10581910556).jpg"
image_path = "images/-_panoramio_-_Basa_Roland-2.jpg"

#reading pattern1
#read and process image
_x = imread(image_path)
x = resize(_x,(224,224))*255.
x  =x.astype(np.uint8)
```

```python
#show image
plt.imshow(x)
plt.show()
#show inference result
show_inference(model,x)
```

```python
#lets visualize the intermediate output (hidden layer images)
intout = intermediate_output(model,x,"block1_conv1",visualize=True,
                             row=6,col=6,cmap="jet")
```

```python
intout = intermediate_output(model,x,"predictions",visualize=True,
                             cmap="jet")
```

```python
#occlusion analysis is finding the important pixel within the image by creating numbers of images with each pixel nullyfied and check how the output changed.
#if 224*224 images, 224*224 images have to be prepared and analyzed so it is very costly and not preferred.
#so sacrificing the final output resolution, we will occulude every XX pixel by the argv skip_pixel and forcely put back to the original resolution
#by resizing.
#another way of occulusion is enlarging the occuluding area (ex. 4x4) but makes the output shape complicated not writing the codes for now.

vals = pixel_wise_occlusion(lmodel,x,skip_pixel=32)
plt.imshow(x)
plt.imshow(vals,alpha=0.5,cmap="jet")
plt.show()

vals = pixel_wise_occlusion(lmodel,x,skip_pixel=16)
plt.imshow(x)
plt.imshow(vals,alpha=0.5,cmap="jet")
plt.show()

vals = pixel_wise_occlusion(lmodel,x,skip_pixel=8)
plt.imshow(x)
plt.imshow(vals,alpha=0.5,cmap="jet")
plt.show()
```

```python
#create a model with guided relu.

gmodel = build_guided_model(model)
```

```python
#The magnitude per pixel, taking the maximum magnitude over the three color channels, as in [2], Section 3.1
#The positive magnitudes (positively correlated with the output), keeping color information
#The negative magnitudes (negatively correlated with the output), keeping color information

gb = guided_backprop(gmodel,x,"predictions",normalize="img_standard")
plt.imshow(gb)
plt.show()

gb = guided_backprop(gmodel,x,"predictions",normalize="abs")
plt.imshow(gb,cmap="jet")
plt.show()

gb = guided_backprop(gmodel,x,"predictions",normalize="pos")
plt.imshow(gb)
plt.show()

gb = guided_backprop(gmodel,x,"predictions",normalize="neg")
plt.imshow(gb)
plt.show()
```

```python
#get all values in each layer
for layer in model.layers:
    print(layer.name)
    if "InputLayer" in str(layer):
        plt.imshow(x)
        plt.show()
    elif "Conv" in str(layer) \
    or "MaxPool" in str(layer) \
    or "Activation" in str(layer):
        intermediate_output(model,x,layer.name,row=6,col=6,cmap="jet")
    elif "Flatten" in str(layer) \
    or "Dense" in str(layer) \
    or "GlobalAverage" in str(layer):
        intermediate_output(model,x,layer.name,onedim=True,cmap="jet")
```
